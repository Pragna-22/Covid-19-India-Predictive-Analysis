# 1. Loading the dataset

```python
import pandas as pd
Covid_19=pd.read_csv("owid-covid-data (3).csv")
print("Rows,columns =",Covid_19.shape)
Covid_19.head()
```

# 2. To subset only those rows that have "India" in the location column by using python's slicing function. 

```python
indexNames = Covid_19[ Covid_19['location'] != 'India' ].index
Covid_19.drop(indexNames , inplace=True)
print("Rows,columns =",Covid_19.shape)
Covid_19.head()
```

# 3. Handling the missing values

a. If there are null values in continuous numerical column, replace the null values by the mean of that column

b. If there are null values in ordinal numerical column, replace the null values by the mode of that column

c. If there are null values in categorical column, replace the null values by the mode of that column

d. If more than 50% the values in a column are null, then drop that entire column



```python
import numpy as np
Covid_num=Covid_19.select_dtypes(include=[np.float])
Covid_num.head()

```

```python
Covid_num.isnull().sum()
```

```python
(Covid_num.isnull().sum())>(len(Covid_num))/2
```

# Since no columns have not more than 50% of null values we don't drop any of the column

```python
Covid_num.mean()
```

```python
Covid_num.fillna(Covid_num.mean(),inplace=True)
Covid_num.isnull().sum()
```

```python
Covid_processed_dataset=(Covid_19.combine_first(Covid_num))
Covid_processed_dataset.isnull().sum()
```

# 8. Predictive modelling using Linear Reegression and Random Forest Regressor


```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
import seaborn as sns
from sklearn.linear_model import LinearRegression
from datetime import date
features = ['date'] #X-axis
target = covid_final_dataset['total_cases']  #Y-axis
plt.figure(figsize=(12, 12))
for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = covid_final_dataset[col]
    y = target
    plt.scatter(x, y, marker='o')
plt.title(col)
p=[]
for i in range(737425,737601,25):
    p.append(i)
t=[]
for i in range(0,len(p)):
    t.append(date.fromordinal(p[i]))
plt.xticks(p, t, rotation=45, horizontalalignment='right')
plt.xlabel(col)
plt.ylabel('total_cases')
reg=LinearRegression()
y=y.values.reshape(-1,1)
x=x.values.reshape(-1,1)
reg.fit(x,y)
prediction_space=np.linspace(min(x),max(x))
prediction_space=prediction_space.reshape(-1,1)
plt.plot(prediction_space,reg.predict(prediction_space),color='red',linewidth=3)
plt.show()
reg_all=LinearRegression()
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3199,random_state=572)
reg_all.fit(X_train,y_train)
y_pred=reg_all.predict(X_test)
p1=reg_all.score(X_test,y_test)
print("Accuracy of the Linear Regression model: ",p1*100,"%")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
X=covid_.drop(['tests_units','total_cases'],axis=1)#Dropping Tests units as it is a categorical feature 
y=covid_final_dataset['total_cases']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2199,random_state=190)
rf = RandomForestRegressor(n_estimators=190,random_state=100)
rf.fit(X_train, y_train) 
y_pred = rf.predict(X_test)
y_pred = y_pred.astype(int)
p2=accuracy_score(y_test,y_pred)
print("Accuracy of the Random Forest Regression model: ",p2*100,"%")
```

```python
plt.figure(figsize=(12, 12))
importances = pd.Series(data=rf.feature_importances_,
                        index= X_train.columns)


importances_sorted = importances.sort_values()

importances_sorted.plot(kind='barh', color='orange')
plt.title('Features Importances')
plt.show()
```
